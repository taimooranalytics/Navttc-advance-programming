```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.style.use('seaborn-white')
```

```python
ages = [25, 26, 27, 28, 29, 30, 31]

dev = [25000, 80000, 100000, 120000, 150000, 180000, 220000]

py_dev = [45000, 58000, 68900, 75000, 89060, 96650, 125520]

js_dev = [55000, 68000, 95900, 115000, 119060, 155650, 225520]

php_dev = [80000, 90000, 110000, 130000, 150000, 180000, 200000]


```

```python
plt.xkcd()
plt.plot(ages, dev)
plt.plot(ages, py_dev, color='b', marker='o', label='python')
plt.plot(ages, js_dev, marker='*', color='c', label='js')
plt.plot(ages, php_dev, linestyle='--', marker='o', label='php')
plt.ylabel("Median salary (USD)")
plt.xlabel("Median salary (USD) by age")
plt.title("this graph is show salary and age")
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig('salary.png')
plt.show()
```
